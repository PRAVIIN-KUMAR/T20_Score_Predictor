```python
import zipfile
import os

# Define the extraction directory
extraction_path = "/content/t20s_extracted" # Or another desired path

# Create the extraction directory if it doesn't exist
os.makedirs(extraction_path, exist_ok=True)

with zipfile.ZipFile("/content/t20s.zip", 'r') as zip_ref:
    # Extract all contents to the specified directory
    zip_ref.extractall(extraction_path)

# Now list files in the extracted directory
filenames = []
for file in os.listdir('/content/t20s_extracted/t20s'):
  filenames.append(os.path.join('/content/t20s_extracted/t20s',file))

print(filenames)
```

```python
import pandas as pd
from tqdm import tqdm
from yaml import safe_load  # or use json.load for JSON files
import os # Import os module if not already imported

final_dfs = []

for counter, file in enumerate(tqdm(filenames), start=1):
    # Check if the file is a YAML file before processing
    if file.endswith('.yaml'):
        with open(file, 'r') as f:
            data = safe_load(f)

            # Debug: Examine the data structure (remove or comment these lines later)
            # print("Type:", type(data))
            # print("Data:", data)

            # Adjust based on the type of data returned
            if isinstance(data, dict):
                # If the data isn't already in a list format
                # If your data is nested, you might want to specify record_path and meta arguments here
                try:
                    df = pd.json_normalize(data)
                except NotImplementedError:
                    # Alternatively, wrap the dict in a list if it represents a single record
                    df = pd.json_normalize([data])
            elif isinstance(data, list):
                df = pd.json_normalize(data)
            else:
                # This should ideally not be reached if we filter by extension,
                # but it's good for robustness.
                print(f"Skipping file with unexpected data format: {file}")
                continue # Skip to the next file

            # Assign a match id and accumulate the result
            df['match_id'] = counter
            final_dfs.append(df)
    else:
        # Optional: Print a message for files being skipped
        print(f"Skipping non-YAML file: {file}")
        continue # Skip to the next file


final_df = pd.concat(final_dfs, ignore_index=True)
```

```python
import pickle
matches = pickle.load(open('/content/dataset_level1.pkl','rb'))
matches
```

```python
count = 1
all_delivery_dfs = []

for index, row in matches.iterrows():
  if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
    count+=1
    continue
  count+=1
  ball_of_match = []
  batsman = []
  bowler = []
  runs = []
  player_of_dismissed = []
  teams = []
  batting_team = []
  match_id = []
  city = []
  venue = []
  if 'innings' in row and row['innings'] and '1st innings' in row['innings'][0] and 'deliveries' in row['innings'][0]['1st innings']:
      for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
          match_id.append(count)
          batting_team.append(row['innings'][0]['1st innings']['team'])
          teams.append(row['info.teams'])
          ball_of_match.append(key)
          batsman.append(ball[key]['batsman'])
          bowler.append(ball[key]['bowler'])
          runs.append(ball[key]['runs']['total'])
          city.append(row['info.city'])
          venue.append(row['info.venue'])
          try:
            player_of_dismissed.append(ball[key]['wicket']['player_out'])

          except:
            player_of_dismissed.append('0')
      loop_df = pd.DataFrame({
          'match_id':match_id,
          'teams':teams,
          'batting_team':batting_team,
          'ball':ball_of_match,
          'batsman':batsman,
          'bowler':bowler,
          'runs': runs,
          'player_dismissed':player_of_dismissed,
          'city':city,
          'venue':venue

      })
      all_delivery_dfs.append(loop_df)
  else:
      print(f"Skipping match_id {count} due to missing or malformed 'innings' data.")


delivery_df = pd.concat(all_delivery_dfs, ignore_index=True)

```

```python
def bowl(row):
  for team in row['teams']:
    if team != row['batting_team']:
      return team
```

```python
delivery_df['bowling_team'] = delivery_df.apply(bowl, axis=1)
```
